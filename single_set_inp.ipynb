```python
#Implement weather modeling* using the quadratic solution in stages: 
#hard-coding variables keyboard input, read from a file, for a single set of input, multiple sets of inputs.

# Keyboard Input

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# keyboard input
a = float(input("Enter a value: "))
b = float(input("Enter b value: "))
c = float(input("Enter c value: "))

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

# Calculate temperature values using the quadratic solution
temperature_values = quadratic_solution(time_values, a, b, c)

plt.plot(time_values, temperature_values)
plt.title('Weather Modeling using Quadratic Solution (Keyboard Input)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.grid(True)
plt.show()
```
